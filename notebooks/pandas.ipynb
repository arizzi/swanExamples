This example shows how to treat basic data with pandas and how to visualise it (inspiration taken [here](http://pandas.pydata.org)).

```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
plt.figure(); df.plot();
```

```python
plt.figure()
df.ix[5].plot(kind='bar'); plt.axhline(0, color='k')
```
